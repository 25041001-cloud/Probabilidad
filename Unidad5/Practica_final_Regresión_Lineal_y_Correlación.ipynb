##**Carga de librerías**

```python
from scipy.stats import pearsonr
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
import seaborn as sns
```

###**Grupo 1**

####Carga y exploración de los datos

```python
df1 = pd.read_csv("grupo1.csv")
df1.head()
```

```python
df1.describe()
```

```python
#Histograma de las calificaciones
plt.hist(df1["Calificacion"], bins=10, color="skyblue", edgecolor="black")
plt.title("Histograma Calificaciones Grupo 1")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.show()
```

####Modelo de Regresión Lineal

```python
HorasGrupo1 = df1[["HorasEstudio"]]
CalificacionGrupo1 = df1["Calificacion"]

modeloGrupo1 = LinearRegression()
modeloGrupo1.fit(HorasGrupo1, CalificacionGrupo1)

print("Coeficiente de la pendiente (m):", modeloGrupo1.coef_[0])
print("Interpretación de la recta (b):", modeloGrupo1.intercept_)
```

####Visualización de resultados

#####Grafico de dispersión y linea de regresión

```python
# Datos Grupo 1
df1 = pd.read_csv("grupo1.csv")
horas_estudio1 = df1["HorasEstudio"].values.reshape(-1, 1)
calificaciones1 = df1["Calificacion"].values

# Definir tamaño
plt.figure(figsize=(6, 5))

# Crear el diagrama de dispersión con línea de regresión
sns.set(style="whitegrid")
sns.regplot(x=horas_estudio1, y=calificaciones1, scatter=True, color='blue')
plt.title('Diagrama de Dispersion con Línea de Regresión')
plt.xlabel('Horas de estudio')
plt.ylabel('Califiación')
plt.grid(True)
plt.show()
```

####Coeficiente de correlación de Peason

```python
# Datos Grupo 1
df1 = pd.read_csv("grupo1.csv")
horas_estudio1 = df1["HorasEstudio"]
calificaciones1 = df1["Calificacion"]

# Calcular coeficiente de correlación de Pearson
coef_corr1, valor_p1 = pearsonr(horas_estudio1, calificaciones1)

print("Coeficiente de correlación de Pearson (Grupo 1):", coef_corr1)
print("Valor p:", valor_p1)
```

###Interpretación de los resultados Grupo 1

####1. ***¿Explicar cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?***

La pendiente es negativa, lo que podemos interpretar como que más horas de estudio se relacionan con una baja en las calificaciones. Esto significa que los datos no muestran una relación lógica.

####2. ***¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y dirección de la relación entre las variables?***

El valor es cercano a cero y negativo, lo que nos dice que tiene una relación muy débil y sin una dirección clara.

####3. ***¿El modelo de regresión lineal es estadísticamente significativo?***

No, porque el valor p es mayor a 0.05.

####4. ***¿Qué tan confiable es el modelo de regresión para predecir las calificaciones finales de los estudiantes basado en el tiempo de estudio?***

Muy poco confiable, no sirve para predecir las calificaciones a partir del tiempo de estudio.
