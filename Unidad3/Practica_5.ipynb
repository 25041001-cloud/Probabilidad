###**Programa 5:** **Gráfica de distribución de probabilidad discreta**

```python
#====================================================
#Grafica de una distribución de probabilidad discreta
#con media, varianza y desviación estandar
#====================================================

import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos: valores de la variable aleatoria y sus probabilidades
x=np.array([5,6,7,8,9,10])
p=np.array([=0.05,0.10,0.20,0.30,0.25,0.10])

#Calculo del valor esperado (media)
esperanza=np.sum(x*p)

#Calculo de la esperanza del cuadrado
esperanza_cuadrado=np.sum(x**2*p)

#Calculo de la varianza y desviación estandar
varianza=esperanza_cuadrado-esperanza**2
desviacion=np.sqrt(varianza)
```

```python
#Mostrar los resultados en consola
print(f"E(X)={esperanza:.2f}")
print(f"E(X^2)={esperanza_cuadrado:.2f}")
print(f"Varianza={varianza:.2f}")
print(f"Desviación estandar={desviacion:.2f}")

#Crear la gráfica tipo 'stem' (tallos y hojas)
plt.stem(x,p,basefmt=" ",linefmt='b-',marketfmt='bo')
plt.title("Distribución de Probabilidad Discreta\ncon Media y Desviación Estandar")
plt.xlabel("Valores de la variable aleatoria X")
plt.ylabel("P(X)")
plt.grid(True,linestyle='--',alpha=0.6)
```
