##Ejemplo de Distribución de probabilidad en Python

```python
import matplotlib.pyplot as plt
```

```python
#Valores de la variable aleatoria
X=[5,6,7,8,9,10]
p=[0.05,0.10,0.20,0.30,0.35,0.10]

#Verificar suma=1
print("Suma de probabilidades:",sum(p))

#Graficar distribución

plt.stem(X,p)
plt.title("Distribución de Probabilidad")
plt.xlabel("Calificación")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```
