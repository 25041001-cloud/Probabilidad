```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
#Verificar que las probabilidades sumen 1
x=[0,10,20,50]
p=[0.50,0.30,0.15,0.05]

sumaProbabilidades=sum(p)
print("La suma de probabilidades:",sumaProbabilidades)
```

```python
#Graficar la distribución de probabilidad

plt.stem(x,p)
plt.title("Distribución de Probabilidad")
plt.xlabel("Monto")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```

```python
#Calcular valor esperado E(X)
#Inicializar una variable para acumular la suma
esperanza=0

#Recorrer los valores y sus probabilidades
for i in range(len(x)):
  xi = x[i]   #Valor de la variable aleatoria
  pi = p[i]   #Probabilidad de ese valor
  producto=xi*pi    #Multiplicar valor por su probabilidad
  esperanza += producto    #sumar al acumulador

  #Mostrar cada paso
  print(f"x={xi}, p={pi}, x*p={producto:.2f}")

#Mostrar resultado final
print(f"\nEsperanza: {esperanza:.2f}")
```

```python
#Calcular varianza y desviación estandar

#Calcular E(x^2)
esperanza_cuadrado=0
for i in range(len(x)):
  esperanza_cuadrado += (x[i]**2)*p[i]

#Calcular varianza y desviación estandar
varianza=esperanza_cuadrado-esperanza**2
desviación=math.sqrt(varianza)

#Mostrar resultados
print(f"E(X)={esperanza:.2f}")
print(f"E(X^2)={esperanza_cuadrado:.2f}")
print(f"Varianza: {varianza:.2f}")
print(f"Desviación estandar={desviación:.2f}")
```

```python
#Calcular la función de distribución acumulada
F=np.cumsum(p)

#Mostrar resultados
for xi,fi in zip(x,F):
  print(f"F({xi}) = {fi:.2f}")

#Graficar
plt.step(x,F,where="post")
plt.title("Funcion de distribución acumulada (FDA)")
plt.xlabel("Monto")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0,1.05)
plt.show()
```
