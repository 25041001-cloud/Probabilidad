```python
#Importar librerias necesarias
import math as mt
from itertools import permutations
from itertools import product
import numpy as np
```

```python
#Tamaños de grupo y subconjunto
asistentes=["asist 1","asist 2","asist 3","asist 4","asist 5","asist 6",
            "asist 7","asist 8","asist 9","asist 10",]
n=10
r=4

#Se realiza el procedimiento para obtener el numero de permutaciones
Pr=mt.factorial(n)/mt.factorial(n-r)

#Las posibles combinaciones de los 10 asistentes de profesor que hay para las
#4 preguntas
print ("Existen ",Pr," posibles permutaciones sin repetición del total de ",
       n," asistentes "," en grupos de ",r" en ",r)

#mostrar permutaciones en grupos de 4 asistentes por examen
print("Aqui las permutaciones")
permutaciones_asistentes = list(permutations(asistentes,r))

for permutations in permutaciones_asistentes:
    print(permutations)
```
