###**Programa 1:** **Combinaciones y permutaciones**

```python
#Basado en el ejercicio de la pagina 25 de la presentación de la unidad 2

import math as mt #Para utilizar factorial
from itertools import combinations
from itertools import combinations_with_replacement
import numpy as np
```

```python
nombres=["Oscar","Paco","Paty","Laura","Ruben","Luis","Lucy","Alberto","Juan","Danny"]
n = len(nombres)
r = 2 #¿Como agrupar?

Cn = mt.factorial(n)/(mt.factorial(r)*mt.factorial(n-r))

print("Existen ",Cn," posibles combinaciones sin repetición del total de ",n," nombres "," en grupos de ",r," en ",r)

print("Aquí las combinaciones")
combinaciones_nombres = list(combinations(nombres,r))

#Imprime las combinaciones
for combinacion in combinaciones_nombres:
  print(combinacion)
```

##Ejemplo 2 Caracteres alfabeticos y digitos

```python
caracteres = ["A","B","C","D","E","F","0","1","2","3","4","5"]
n = len(caracteres)
r = 10 # ¿Cómo agrupar?

Cn = mt.factorial(n)/mt.factorial(n-r)
print("Existen ",Cn," posibles combinaciones sin repetición del total de ",n,
      " caracteres "," en grupos de ",r," en ",r)

print ("Aquí las combinaciones")
combinaciones_caracteres = list(combinations(caracteres,r))

#Imprime las combinaciones
for combinacion in combinaciones_caracteres:
  print(combinacion)
```
