```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
#Cargar el archivo CDV en una variable llamada covid
covid=pd.read_csv("20-06-30-COVID19MEXICO - codificado.csv",encoding="ISO-8859-1")
```

```python
#Mostramos los primeros 5 registros del archivo
covid.head()
```

```python
#Mostrar la información del archivo.
#Muestra el numero de registros, tipo de datos de la columna, valores nulos, memoria

covid.info()
```

```python
#Mostrar la edad maxima de la base de datos de la columna edad
covid.EDAD.max()
```

```python
#Mostrar la edad minima de la base de datos de la columna edad
covid.EDAD.min()
```

```python
#Dibuja el histograma del atributo "EDAD" con bins=3
#Calcular la media y guardarla en una variable para mostrarle en el histograma
media=covid.EDAD.mean()
plt.pyplot.hist(covid.EDAD,bins=3)
plt.pyplot.title("Histograma EDAD Pruebas COVID")
plt.pyplot.axvline(media,color='y',linestyle='dashed',linewidth=2, label=f'Media: {media:.2f}')
plt.pyplot.xlabel("Edad")
plt.pyplot.ylabel("Frecuencia")
```

```python
#Dibuja el histograma de la columna sexo,
#esto permite comparar cuantos hay de cada grupo
plt.pyplot.hist(covid.SEXO,bins=3)

#Establece las etiquetas x/y y muestra el titulo
plt.pyplot.title("Histograma SEXO Pruebas COVID")
plt.pyplot.xlabel("Sexo")
plt.pyplot.ylabel("Frecuencia")
```

```python
#Dibuja el histograma del atributo "Tipo Paciente", esto permite
#comparar cuantos hay de cad grupo
plt.pyplot.hist(covid.TIPO_PACIENTE,bins=3)

#establece las etiquetas x/y y muestra el titulo
plt.pyplot.title("Histograma tipos de pacientes en Pruebas COVID")
plt.pyplot.xlabel("Tipo Paciente")
plt.pyplot.ylabel("Frecuencia")
```

```python
#Crear un diagrama de cajas para visualizar la distribución
#de los rangos de edad por sexo
sns.boxplot(x="SEXO",y="EDAD",data=covid)
```

```python
#Crear un diagrama de cajas para visualizar la distribución
#de los rangos de edad por tipo de paciente
sns.boxplot(x="TIPO_PACIENTE",y="EDAD",data=covid)
```

```python
#Crear un diagrama de cajas para visualizar la distribución
#de los rangos de edad por tipo de paciente
sns.boxplot(x="RESULTADO",y="EDAD",data=covid)
```
