###**Problemas de Distribución Binomial y Poisson**

#PROBLEMA 1

Es tipo Poisson porque los datos representan el numero de personas que ingresan a la unidad de cuidados intensivos en un día, los eventos ocurren en un intervalo de tiempo fijo, que vendría siendo un día.

```python
from scipy.stats import poisson

# Parámetro de la distribución
media_ingresos_dia = 5

# a) Probabilidad de que ingresen 2 personas
p2 = poisson.pmf(2, media_ingresos_dia)
print("Probabilidad de que hayan 2 ingresos en un día:", p2)

# b) Probabilidad de que exceda 10
p_mas_10 = 1 - poisson.cdf(10, media_ingresos_dia)
print("Probabilidad de que hayan más de 10 ingresos en un día:", p_mas_10)
```

No es probable que x exceda 10, pues la probabilidad es de 1.37%, la cual es muy baja para que suceda en un contexto más realista

#PROBLEMA 2

Es del tipo binomial porque en los datos hay 1000 motores, los cuales solo pueden tener dos posibles resultados, que sean satisfactorios o no satisfactorios (exito o fracaso) y la probabilidad de exito es igual en todos los eventos, en el problema se plantea una probabilidad del 0.001 para todos los motores.

```python
from scipy.stats import binom

# Parámetros
motores = 1000
probabilidad_de_exito = 0.001

# 1.- Probabilidad de que ningun motor saldrá defectuoso
probabilidad_de_0 = binom.pmf(0, motores, probabilidad_de_exito)
print("Probabilidad de 0 defectuosos:", probabilidad_de_0)

# b) Probabilidad de que al menos tres salgan defectuosos
probabilidad_de_3 = binom.pmf(3, motores, probabilidad_de_exito)
print("Probabilidad de 3 defectuosos:", probabilidad_de_3)

# c) Cuatro defectuosos
probabilidad_de_4 = binom.pmf(4, motores, probabilidad_de_exito)
print("Probabilidad de 4 defectuosos:", probabilidad_de_4)
```
