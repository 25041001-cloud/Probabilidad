###**PROBLEMA 1:** **Estudiantes que pesan más de 68 y menos de 72**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
plt.figure(figsize=(6, 4))

# Parámetros de la distribución normal
media = 70   # Media
desv_std = 4 # Desviación estándar
total_estudiantes = 500

# Límites de peso
peso_min = 68
peso_max = 72

# Probabilidad de pesar menos de 72 kg
prob_menos_72kg = norm.cdf(peso_max, media, desv_std)

# Probabilidad de pesar menos de 68 kg
prob_menos_68kg = norm.cdf(peso_min, media, desv_std)

# Probabilidad de estar entre 68 y 72 kg
prob_entre_68_72kg = prob_menos_72kg - prob_menos_68kg

# Número de estudiantes que pesan entre 68 y 72 kg
estudiantes_entre_68_72kg = total_estudiantes * prob_entre_68_72kg

print("Probabilidad de que un estudiante pese entre 68 y 72 kg:", prob_entre_68_72kg)
print("Número de estudiantes que pesan entre 68 y 72 kg:", estudiantes_entre_68_72kg)

# Calcular la distribución normal
x = np.linspace(media - 4*desv_std, media + 4*desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Graficar la distribución normal
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_min) & (x <= peso_max), color='skyblue', alpha=0.5)
plt.title(r'Probabilidad de que un estudiante pese entre 68 y 72 kg. $\mu=70$, $\sigma=4$, n=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
